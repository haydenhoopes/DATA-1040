```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 257
pd.options.display.max_colwidth = None
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('shark_attacks.csv', low_memory=False)
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df = df.drop(columns=['href formula', 'pdf', 'Unnamed: 23', 'Unnamed: 9'])
```

```python
df.head(5)
```

```python
temp_df = df['Date'].str.extract('(\d\d\d\d)')
df['Year'] = temp_df[0]
```

```python
temp_df = df['Date'].str.extract('(\d?\d)\W(\w{3,4})\W*(\d\d\d\d)')
df[['Day', 'Month']] = temp_df[[0, 1]]
```

```python
temp_df = df['Case Number'].str.extract('(\d\d\d\d).(\d\d).(\d\d)')
df['Day'] = df['Day'].fillna(temp_df[2])
df['Month'] = df['Month'].fillna(temp_df[1])
```

```python
df.head()
```

```python
df[['Hour', 'Minutes']] = df['Time'].str.extract('(\d\d)h(\d\d)')
```

```python
df['Hour'] = df['Hour'].fillna(df['Time'])
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Year', 'Day'])
```

```python
month_map = {
    '01': 'Jan',
    '02': 'Feb',
    '03': 'Mar',
    '04': 'Apr',
    '05': 'May',
    '06': 'Jun',
    '07': 'Jul',
    '08': 'Aug',
    '09': 'Sep',
    '10': 'Oct',
    '11': 'Nov',
    '12': 'Dec',
    'July': 'Jul',
    'Sept': 'Sep',
    'Decp': 'Dec',
    '00': np.NaN
}

df['Month'] = df['Month'].replace(month_map)
```

```python
df['Day'] = df['Day'].apply(lambda x : "0" + str(x) if len(str(x)) == 1 else x)
```

```python
df['Day'] = df['Day'].replace({'00': np.NaN})
```

```python
df['Type'] = df['Type'].replace({'Boat': 'Boating', 'Boatomg':'Boating', 'Unverified':'Unconfirmed', 'Under investigation':'Unconfirmed'})
```
