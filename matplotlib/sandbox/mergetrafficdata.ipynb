```python
import pandas as pd
```

```python
meta_df = pd.read_csv('Utah Traffic MetaData.csv')
df = pd.read_csv('Utah Traffic.csv')
```

```python
meta_df.head()
```

```python
df.head()
```

```python
df = df.melt(id_vars='Unnamed: 0', value_vars=df.columns[1:], var_name='station_id', value_name='traffic').rename(columns={'Unnamed: 0': 'Datetime'})
```

```python
df['station_id'] = df['station_id'].astype('int64')
```

```python
df.merge(right=meta_df, on='station_id').to_csv('salt_lake_traffic.csv', index=False)
```
