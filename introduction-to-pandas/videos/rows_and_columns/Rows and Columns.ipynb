```python
def pokemonType(row):
    if row['type2'] == "None":
        print(f"{row['pokename']} is a {row['type1']} pokemon")
    else:
        print(f"{row['pokename']} is a {row['type1']} and {row['type2']} pokemon")
```

```python
df.apply(pokemonType, axis=1)
```

```python
def greaterThanAverageSpeed(x):
    average_speed = df['speed'].mean()
    return x > average_speed
    
df['speed'].apply(greaterThanAverageSpeed)
```
