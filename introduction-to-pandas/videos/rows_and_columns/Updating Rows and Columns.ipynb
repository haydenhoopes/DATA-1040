# 2.6 Updating Rows and Columns

```python
import pandas as pd

df = pd.read_csv("pokemon.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns[0] = 'Trainer ID'
```

```python
df.rename(columns={'trainerID':'Trainer ID'}, inplace=True)
```

```python
df.head()
```

```python
def pokemonType(row):
    if row['type2'] == 'None':      
        print(f"{row['pokename']} is a {row['type1']} pokemon")
    else:
        print(f"{row['pokename']} is a {row['type1']} and a {row['type2']} pokemon")
```

```python
df.apply(pokemonType, axis=1)
```

```python
def greaterThanAverageSpeed(x):
    average_speed = df['speed'].mean()
    return x > average_speed

df['speed'].apply(greaterThanAverageSpeed)
```
