# 2.5 Filtering
This video discusses how to create dataframe filters using Pandas.

```python
import pandas as pd
df = pd.read_csv("pokemon.csv")

df.head()
```

```python
df['type1'] == 'Fire'
```

```python
df[ df['type1'] == 'Fire' ]
```

```python
filt = df['hp'] > 100
```

```python
df[filt]
```

```python
df.loc[df['hp'] > 100, 'pokename']
```

```python
~ not
& and
| or
```

```python
filt = (df['hp'] > 100) & (df['type1'] == 'Fire')
```

```python
df.loc[filt]
```

```python
filt = (df['hp'] < 50) | (df['hp'] > 200)
```

```python
df.loc[filt]
```

```python
filt = ~(df['pokename'] == 'Pikachu') & (df['type1'] == 'Electric')
```

```python
df.loc[filt, 'pokename'].value_counts()
```

```python
water_df = df.loc[ df['type1'] == 'Water' ]
```

```python
water_df.loc[ water_df['type2'] == 'Fire']
```
