### The `.str` accessor object

```python
df['text'].str.len()
```

### Examples of `.str` methods

```python
filt = df['text'].str.contains('Obama')
```

```python
df.loc[ filt ] 
```

```python
filt = df['created_at'].str.contains('12')
```

```python
df.loc[ filt ]
```

```python
df['month'] = df['created_at'].str.split('/').str[1]
```

```python
df.loc[ df['month'] == '12' ]
```
