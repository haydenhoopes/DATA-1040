```python
df.loc[ df['user'] == 'SenatorHassan', 'state'] = 'NH'
```

```python
df['state'].fillna('UT', inplace=True)
```

```python
df['null_state'] = df['state'].isna()
def fix_null_states(row):
    if row['null_state']:
        return df.loc[ df['user'] == row['user']]['state'].mode()[0]
    else:
        return row['state']
```

```python
df['state'] = df.apply(fix_null_states, axis=1)
```

```python
df['state'].isna()
df.loc[ df['state'].isna() ]
```

```python
import numpy as np
q1 = np.quantile(df['replies'], .25)
q3 = np.quantile(df['replies'], .75)
iqr = q3 - q1

def is_outlier(row):
    return row > q3 + (1.5 * iqr)

df['replies'].apply(is_outlier).sum()
```

```python
letters = {}
def count_letters(text):
    for letter in text:
        if letter in letters:
            letters[letter] += 1
        else:
            letters[letter] = 1
    
```
