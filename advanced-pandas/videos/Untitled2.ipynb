```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('senator_tweets.csv')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 8), dpi=160)
plt.scatter(df['replies'], df['retweets'], alpha=0.2)
plt.xlabel('replies')
plt.ylabel('retweets')
plt.title('Relationship between Replies and Retweets')
plt.show()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df.set_index('created_at', inplace=True)
```

```python
utah_retweets = df.loc[df['state'] == 'UT', 'retweets'].resample('M').mean()

utah_x = utah_retweets.index
utah_y = utah_retweets
```

```python
all_retweets = df['retweets'].resample('M').mean()
all_x = all_retweets.index
all_y = all_retweets
```

```python
utah_retweets = df.loc[df['state'] == 'UT', 'retweets'].resample('M').mean()
utah_x = utah_retweets.index
utah_y = utah_retweets

all_retweets = df['retweets'].resample('M').mean()
all_x = all_retweets.index
all_y = all_retweets

import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4), dpi=160)
plt.plot(all_x, all_y, label="US", alpha=.4)
plt.plot(utah_x, utah_y, label="Utah")
plt.xlabel('Year')
plt.ylabel('Average Retweets')
plt.title('Average Retweets per Month')
plt.legend(loc='upper left')
plt.show()
```
