# 3.6 Date Formatting

```python
import pandas as pd
df = pd.read_csv('senator_tweets.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df['created_at'].dt.normalize()
```

```python
filt = df['created_at'].dt.normalize() == this_date
df.loc[filt]
```

```python
from datetime import datetime
this_date = datetime(2017, 10, 19)
```

```python
amy_df = df.loc[ df['user'] == 'amyklobuchar']
```

```python
first_tweet = amy_df['created_at'].min()
most_recent_tweet = amy_df['created_at'].max()
```

```python
twitter_time = most_recent_tweet - first_tweet
```

```python
twitter_time.days / 365
```

```python
df['year'] = df['created_at'].dt.year
```

```python
df.groupby('year').agg({'retweets': 'mean'})
```

```python
df['month'] = df['created_at'].dt.month
```

```python
df.groupby(['year', 'month']).agg({'retweets': 'mean'})
```
