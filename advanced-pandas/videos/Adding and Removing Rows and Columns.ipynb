```python
new_tweet = {
    'text': ['I love chickens!'],
    'url': ['https://en.wikipedia.org/wiki/Chicken'],
    'replies': [10],
    'retweets': [27],
    'favorites': [133],
    'user': ['senator_hayden'],
    'state': ['UT']
}
```

```python
new_tweet_df = pd.DataFrame(new_tweet)
```

```python
df = pd.concat([df, new_tweet_df], ignore_index=True)
```
