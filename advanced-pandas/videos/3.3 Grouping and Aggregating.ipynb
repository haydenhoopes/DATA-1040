# 3.3 Grouping and Aggregating

```python
import pandas as pd
df = pd.read_csv("senator_tweets.csv")
```

```python
df.head()
```

```python
df['replies'].max()
```

```python
df.groupby('party').count()
```

```python
df.groupby('party').mean()
```

```python
df.groupby('party').agg({'replies': ['mean', 'median'], 'retweets': ['mean', 'max'], 'favorites': ['mean', 'std']})
```

```python
df.groupby(['state', 'party']).agg({'replies': 'mean', 'retweets':'mean'})
```
