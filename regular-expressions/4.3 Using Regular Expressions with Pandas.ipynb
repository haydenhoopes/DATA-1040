# 4.3 Using Regular Expressions with Pandas

```python
import pandas as pd
df = pd.read_csv('./data/books.csv')
```

```python
df.head()
```

```python
filt = df['favorite_quote'].str.contains('\shappy|happiness')
df.loc[filt]
```

```python
filt = df['description'].str.contains('b.+$')
df.loc[filt]
```

```python
df['pages'] = df['description'].str.extract('with\s(\d+)\spages')[0]
```

```python
df.head()
```

```python
df['title'] = df['One_line_review'].str.extract('^(([A-Z0-9][\w%\'-]*\s)*)')[0]
```
